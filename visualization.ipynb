# Library

```python
import pandas as pd
import matplotlib as plt
import re
```

# LWM

```python
file_path = "LWM_epoch1000.txt"

LWM_train_pattern = re.compile(r"TrainLoss:\s*([0-9.]+)")
LWM_val_pattern = re.compile(r"ValLoss:\s*([0-9.]+)")
LWM_NMSE_dB_pattern = re.compile(r"Val NMSE_dB:\s*([-0-9.]+)")

LWM_train_loss = []
LWM_val_loss = []
LWM_NMSE_dB = []

# read file
with open(file_path, "r") as f:
    for line in f:
        train_loss = LWM_train_pattern.search(line)
        val_loss = LWM_val_pattern.search(line)
        NMSE_dB = LWM_NMSE_dB_pattern.search(line)

        if train_loss and val_loss and NMSE_dB:
            LWM_train_loss.append(float(train_loss.group(1)))
            LWM_val_loss.append(float(val_loss.group(1)))
            LWM_NMSE_dB.append(float(NMSE_dB.group(1)))

print("Train_loss : ",LWM_train_loss[:10])
print("Val loss: ",LWM_val_loss[:10])
print("NMSE_dB : ",LWM_NMSE_dB[:10])
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(LWM_train_loss, label='Train Loss')
plt.plot(LWM_val_loss, label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('LWM Training and Validation Loss')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(LWM_NMSE_dB, label='Val NMSE (dB)', color='green')
plt.xlabel('Epoch')
plt.ylabel('NMSE (dB)')
plt.title('Validation NMSE (dB) over Epochs')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
